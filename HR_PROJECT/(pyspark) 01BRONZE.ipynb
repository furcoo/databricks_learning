```python
from delta.tables import *
from pyspark.sql.functions import col, current_timestamp

# Load and write employees data
df = spark.read.format("csv").option("header", True).load("/Volumes/uc_data001/uc_schema001/volume001/Employees/employees1*.csv")
df.write.mode("overwrite").saveAsTable("uc_data001.uc_schema001.raw_hr_employees")

# Load and write countries data
df = spark.read.format("csv").option("header", True).load("/Volumes/uc_data001/uc_schema001/volume001/Employees/countries*.csv")
df.write.mode("overwrite").saveAsTable("uc_data001.uc_schema001.raw_hr_employees_countries")

# Load and write departments data
df = spark.read.format("csv").option("header", True).load("/Volumes/uc_data001/uc_schema001/volume001/Employees/departments*.csv")
df.write.mode("overwrite").saveAsTable("uc_data001.uc_schema001.raw_hr_employees_departments")

# Load and write locations data
df = spark.read.format("csv").option("header", True).load("/Volumes/uc_data001/uc_schema001/volume001/Employees/locations*.csv")
df.write.mode("overwrite").saveAsTable("uc_data001.uc_schema001.raw_hr_employees_locations")
```
