#项目：HR_PROJECT
##这是用来定义所有的表对象的DDL语句
https://learn.microsoft.com/zh-cn/azure/databricks/sql/get-started/data-warehousing-concepts

https://learn.microsoft.com/zh-cn/azure/databricks/lakehouse/medallion

##this is bronze data from source


```python

%sql
CREATE OR REPLACE TABLE uc_data001.uc_schema001.raw_hr_employees (
  EMPLOYEE_ID STRING,
  FIRST_NAME STRING,
  LAST_NAME STRING,
  EMAIL STRING,
  PHONE_NUMBER STRING,
  HIRE_DATE STRING,
  JOB_ID STRING,
  SALARY STRING,
  MANAGER_ID STRING,
  DEPARTMENT_ID STRING)
USING delta;

CREATE OR REPLACE TABLE uc_data001.uc_schema001.raw_hr_employees_countries (
  COUNTRY_ID STRING,
  COUNTRY_NAME STRING)
USING delta;

--接入层的数据不能有INT  否则插入就会报错，或者提前做格式转换
CREATE OR REPLACE TABLE uc_data001.uc_schema001.raw_hr_employees_departments (
  DEPARTMENT_ID STRING,
  DEPARTMENT_NAME STRING,
  MANAGER_ID STRING,
  LOCATION_ID STRING)
USING delta;

CREATE OR REPLACE TABLE uc_data001.uc_schema001.raw_hr_employees_locations (
  LOCATION_ID STRING,
  STREET_ADDRESS STRING,
  POSTAL_CODE STRING,
  CITY STRING,
  STATE_PROVINCE STRING,
  COUNTRY_ID STRING)
USING delta;
```

##this is silver layer data

```python

%sql
CREATE OR REPLACE TABLE uc_data001.uc_schema001.fct_hr_employees (
  EMPLOYEE_ID STRING,
  FIRST_NAME STRING,
  LAST_NAME STRING,
  EMAIL STRING,
  PHONE_NUMBER STRING,
  HIRE_DATE TIMESTAMP_NTZ,
  JOB_ID STRING,
  SALARY STRING,
  MANAGER_ID STRING,
  DEPARTMENT_ID STRING,
  DB_CREATED_DATE TIMESTAMP_NTZ,
  DB_UPDATED_DATE TIMESTAMP_NTZ
  )
USING delta;

CREATE OR REPLACE TABLE uc_data001.uc_schema001.dim_hr_employees_countries (
  COUNTRY_ID STRING,
  COUNTRY_NAME STRING,
  DB_CREATED_DATE TIMESTAMP_NTZ,
  DB_UPDATED_DATE TIMESTAMP_NTZ)
USING delta;

CREATE OR REPLACE TABLE uc_data001.uc_schema001.dim_hr_employees_departments (
  DEPARTMENT_ID BIGINT,
  DEPARTMENT_NAME STRING,
  MANAGER_ID BIGINT,
  LOCATION_ID BIGINT,
  DB_CREATED_DATE TIMESTAMP_NTZ,
  DB_UPDATED_DATE TIMESTAMP_NTZ)
USING delta;

CREATE OR REPLACE TABLE uc_data001.uc_schema001.dim_hr_employees_locations (
  LOCATION_ID BIGINT,
  STREET_ADDRESS STRING,
  POSTAL_CODE STRING,
  CITY STRING,
  STATE_PROVINCE STRING,
  COUNTRY_ID STRING,
  DB_CREATED_DATE TIMESTAMP_NTZ,
  DB_UPDATED_DATE TIMESTAMP_NTZ)
USING delta;
```
